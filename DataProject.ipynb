```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/S-JG/DAP/main/pixar_rotten_tomatoes.csv")

#연도별 오름차순 정렬
dfA = df.sort_values('release_year') 

df.head()

```

**Trends in ratings of Pixar movies up to 2021**

```python
#연도별 오름차순 정렬
dfA = df.sort_values('release_year') 

#영화, 토마토평점, 일반평점 리스트로 추출
x = dfA['movie'].tolist()
y1 = dfA['tomatometer_rating'].tolist()
y2 = dfA['audience_rating'].tolist()

#그래프 그리기
plt.figure(figsize=(80,8))
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')

#그래프 제목 설정 
title_font = {
              'fontsize':30,
              'fontweight':'bold'
              }

plt.title("Pixar Movies' Ratings until 2021", fontdict=title_font)


#그래프 눈금 설정
font1 = {
         'color': 'black',
         'weight': 'bold',
         'size': 30
         }

font2 = {
         'color': 'black',
         'weight': 'bold',
         'size': 30
         }


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlabel('Movies', fontdict=font1)
plt.ylabel('Ratings', fontdict=font2)

#그래프 범례설정
plt.legend(['tomatometer_rating', 'audience_rating'], loc='lower right', fontsize=30)

#그래프 격자무늬 설정
plt.grid(True)

plt.show()
```

**Regression: Ratings**

```python
from sklearn.linear_model import LinearRegression

X = dfA['release_year']
y1 = dfA['tomatometer_rating']
y2 = dfA['audience_rating']

line_filter_1 = LinearRegression()
line_filter_1.fit(X.values.reshape(-1,1), y1)

line_filter_2 = LinearRegression()
line_filter_2.fit(X.values.reshape(-1,1), y2)

plt.figure(figsize=(50,8))

#그래프 눈금 설정
font1 = {
         'color': 'black',
         'weight': 'bold',
         'size': 30
         }

font2 = {
         'color': 'black',
         'weight': 'bold',
         'size': 30
         }

#토마토
plt.subplot(1,2,1)
plt.plot(X, y1, 'r')
plt.plot(X,line_filter_1.predict(X.values.reshape(-1,1)))

plt.grid(True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlabel('Realease year', fontdict=font1)
plt.ylabel('Tomatometer_Ratings', fontdict=font2)

#관객 평가
plt.subplot(1,2,2)
plt.plot(X, y2, 'b')
plt.plot(X,line_filter_2.predict(X.values.reshape(-1,1)),)

plt.grid(True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.xlabel('Realease year', fontdict=font1)
plt.ylabel('Audience_Ratings', fontdict=font2)


plt.show()
```
